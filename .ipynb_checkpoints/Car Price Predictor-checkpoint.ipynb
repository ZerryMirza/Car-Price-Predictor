#### Data Preprocessing

Data Type

```python
# year has not int values like zest, sales etc. should be converted to int and drop the rest
df = df[df['year'].str.isnumeric()]
```

```python
df['year'] = df['year'].astype(int)
```

```python
# Null Values
df.isnull().sum()
```

```python
df['Price'].head()
# Contains "Ask For Price"
```

```python
df = df[df['Price'] != "Ask For Price"]
```

```python
df['Price'].head() # dtype = Object
```

```python
df['Price'] = df['Price'].str.replace(",",'').astype(float)
```

```python
df['kms_driven'].value_counts()
```

```python
df['fuel_type']
```

```python
# Find the relation of "KMS Driven" with other columns
```

```python
plt.figure(figsize=(32,8))
ax = sns.countplot(df['kms_driven'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
sns.boxplot(df['kms_driven'], df['company'])
```

```python
# Fetching numeric value from kms_driven column
```

```python
df['kms_driven_float'] = df['kms_driven'].str.split(' ', expand=True)[0]
```

```python
df['kms_driven_float'] = df['kms_driven_float'].str.replace(',','')
```

```python
df['kms_driven_float'] = pd.to_numeric(df['kms_driven_float'], errors='coerce')
```

```python
df['kms_driven_float']
```

```python
df
```

```python
plt.figure(figsize=(24,8))
ax = sns.boxplot(data=df, y='kms_driven_float', x='company')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
# see the company values where kms_driven is null
```

```python
df[df['kms_driven_float'].isnull()]['company'].value_counts()
```
